```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from IPython.display import display, HTML
```

```python
data = pd.read_csv('data/winequality-red.csv', sep=';')
data = data.iloc[:10, :]
display(data)
```

# Calculate the Min Max Scaling of the Data

```python
# Calculating the min max scaling manually
datamm = (data - data.min()) / (data.max() -data.min())
```

```python
#Calculating min max scaling using sklearn
minmax_scale = preprocessing.MinMaxScaler().fit(data)
data_minmax = minmax_scale.transform(data)
data_minmax = pd.DataFrame(data_minmax)
data_minmax.columns = data.columns
display(data_minmax)
```

# Calculate the Z score of the data

```python
# calculating the z score manually
# print(data.mean(), data.std(), data)
data_zscore = (data - data.mean()) / data.std()
#display(data_zscore)
```

```python
std_scale = preprocessing.StandardScaler().fit(data)
data_zscore = pd.DataFrame(std_scale.transform(data))
data_zscore.columns = data.columns
display(data_zscore)
```

# Calcualte distance of data

```python
from scipy.spatial import distance
```

## Manhatten Distance

```python
# calculate the pairwise distance using the manhatten distance metic
d = pd.Series(distance.pdist(data, 'cityblock'))
dsq = pd.DataFrame(distance.squareform(d))
np.fill_diagonal(dsq.values, np.nan)
display(dsq)
```

```python
distances = pd.DataFrame([dsq.idxmin(axis=0), dsq.min(axis=0), dsq.idxmax(axis=0), dsq.max(axis=0)]).T
distances.columns = ['nearest_manhatten_point', 'near_manhatten_dist_val','far_manhatten_point', 'far_manhatten_dist_val']
display(distances)
```
